```python
df = spark.read.table("population_metrics.default.countries_population")
```

```python
from pyspark.sql.functions import round, greatest, least, length, lower, upper, concat, lit, concat_ws, to_date, to_timestamp, date_format
```

```python

df2 = df.select(df.country_id, df.name, df.population, round(df.population * 1.2, 1).alias("2028_population_forecast"))

df2.display()
```

```python
df2.select("name", "population","2028_population_forecast",least("population", "2028_population_forecast").alias("least_population"), greatest("population", "2028_population_forecast").alias("greatest_population")).display()
```

```python
df.select("name", length("name").alias("name_length"), lower("name"), upper("name")).display()
```

```python
df.select("name", "capital", concat("name", lit(", "),"capital"), concat_ws(", ", "name", "capital")).display()
```

```python
schema = "name string, hire_date string"

# Create a minimal DataFrame with name and hire_date
data = [
    ("Alice", "15-06-2020"),
    ("Bob", "25-09-2018"),
    ("Charlie", "05-12-2022")
    ]

df_date = spark.createDataFrame(data, schema)
df_date.display()
```

```python
df_date = df_date.select("name", "hire_date", to_date("hire_date", "dd-MM-yyyy").alias("date"), to_timestamp("hire_date", "dd-MM-yyyy").alias("timestamp"))
df_date.display()
```
