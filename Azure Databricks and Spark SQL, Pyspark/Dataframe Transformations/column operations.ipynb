```python
df = spark.read.table("population_metrics.default.countries_population")
```

```python
df.select("country_id", "name","country_code").orderBy("country_id").display()
```

```python
df.select(df['country_id'], df['name'], df['country_code']).display()
```

```python
df.select(df.country_id, df.name, df.country_code).display()
```

```python
df.select(df.country_id.alias("country_identifier"), df.name.alias("country_name"), df.country_code).display()
```

```python
df.select(df["country_id"].alias("id"), df['name'].alias('Country_name')).display()
```

```python
from pyspark.sql.functions import col
```

```python
df.select(col("country_id").alias("id"), col("name").alias("Country_name")).display()
```

```python
df.groupBy(df.name).count().filter("count>1").display()
```

```python
%sql
select country_id, count(1) from population_metrics.default.countries_population
group by country_id
having count(1) > 1
```

```python
from pyspark.sql.functions import col, upper
```

```python
df.select(
    col("country_id"),
    upper(col("name")).alias("country_name"),
    col("country_code"),
    col("population"),
    col("area_km2"),
    (col("population")/col("area_km2")).alias("density")
    ).display()
```

```python
df.selectExpr(
    "country_id",
    "name as country_name",
    "country_code",
    "population",
    "area_km2",
    "(population/area_km2) as density"     
    ).display()
```

```python
df.select(df.country_id, upper(df.name).alias("country_name"), df.population, df.area_km2, (df.population/df.area_km2).alias("density")).display()
```

```python
df\
.withColumnRenamed("name", "country_name")\
.withColumnRenamed("capital", "country_capital")\
.display()
```

```python
df.withColumnsRenamed({

    "name": "country_name",
    "capital": "country_capital"
    
    }).display()
```

```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import col
```

```python
df2 = spark.read.table("population_metrics.default.countries_population_tb")
```

```python
df2_1 = df2.select(
    col("country_id").cast(StringType()),
    upper(col("name")).alias("country_name"),
    col("nationality"),
    col("country_code"),
    col("population").cast(StringType()),
    col("area_km2").cast(StringType()),
    col("region_id").cast(StringType()),
    col("sub_region_id").cast(StringType())
)
```

```python
df2_1.display()
```

```python
df2_2 = df2.select(
    df2.country_id.cast(StringType()),
    df2.name.alias("country_name"),
    df2.nationality,
    df2.country_code,
    df2.population.cast(StringType()),
    df2.area_km2.cast(StringType()),
    df2.region_id.cast(StringType()),
    df2.sub_region_id.cast(StringType())
    
)
df2_2.display()
```
