# Removing Duplicates

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Define Schema
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("department", StringType(), True)
])

# Sample Data (with Duplicates)
data = [
    (1, "Alice", "HR"),
    (2, "Bob", "IT"),
    (3, "Charlie", "Finance"),
    (1, "Alice", "HR"),  # Duplicate row
    (2, "Bob", "IT"),    # Duplicate row
    (4, "David", "HR"),
    (3, "Charlie", "Finance"),  # Duplicate row
    (5, "Alice", "Finance"),  # Same name, different department
    (6, "Bob", "HR")  # Same name, different department
]

# Create DataFrame
df = spark.createDataFrame(data, schema)

df.show()
```

```python
df.dropDuplicates().display()
```

```python
df.dropDuplicates(["name"]).display()
```

```python
df.dropDuplicates(["name", "department"]).display()
```

# Sorting and Limiting

```python
df = spark.read.table("population_metrics.default.countries_metrics")
```

```python
from pyspark.sql.window import Windowfrom pyspark.sql.functions import rank

window_spec = Window.orderBy('column_name')
df.withColumn('rank', rank().over(window_spec)).display()
```
