```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

schema = StructType([
    StructField('country_id', IntegerType(), False),
    StructField('name', StringType(), True),
    StructField('nationality', StringType(), True),
    StructField('country_code', StringType(), True),
    StructField('iso_alpha2', StringType(), True),
    StructField('capital', StringType(), True),
    StructField('population', IntegerType(), True),
    StructField('area_km2', DoubleType(), True),
    StructField('region_id', IntegerType(), True),
    StructField('sub_region_id', IntegerType(), True)
])

df = spark.read.format('csv').schema(schema).options(header = True).load('/Volumes/population_metrics/landing/datasets/countries_dataset/countries_dataset/csv_data/countries_population/countries_population.csv')

df.write.mode('overwrite').saveAsTable('population_metrics.default.countries_population_tb')
```

```python
%sql

create table if not exists population_metrics.default.countries_population_tb2 (
country_id	int not null comment 'Country ID',
name	string comment 'Country name',
nationality	string comment 'nationality',
country_code	string comment 'country_code',
iso_alpha2	string comment 'Country iso_alpha2',
capital	string comment 'Country capital',
population	int comment 'Country population',
area_km2	double comment 'Country area_km2',
region_id	int comment 'Country region_id',
sub_region_id	int comment 'Country sub_region_id',
constraint country_pk primary key (country_id)
)
using delta
CLUSTER BY auto
comment "Countries Population Metrics"
TBLPROPERTIES (
delta.autoOptimize.optimizeWrite = true,
delta.autoOptimize.autoCompact = true,
delta.enableIcebergCompatV2 = true,
delta.universalFormat.enabledFormats = 'iceberg'
)
```
